```python
indice=[]
    
phrase= "la  du 15 juillet et la  de 2000 ne sont pas si importantes"

doc=nlp(phrase)
    
def get_annotation_balise(phrase):
    doc=nlp(phrase)
    indice=[]
    phrase= doc
    liste_doc=["décret-loi","arrêté","loi","texte", "sujétion","schéma", "plan",
         "décret", "édit", "programme", "Décr","réglement", "charte",
          "ordonnance","article"
         ]
    
    for p in range(len(doc)):
        for mot in liste_doc:
            if doc[p].lemma_ == mot and doc[p].pos_=="NOUN":
                indice.append(p)
         

        if len(indice)>=1:
            l= indice[0]
            span = doc[doc[indice[0]].left_edge.i : doc[indice[0]].right_edge.i+1]
            if span.text != ' ' :
        
                debut= doc[indice[0]].left_edge.i
                fin= doc[indice[0]].right_edge.i+1
        
                phrase= f"{doc[:debut]}<doc>{doc[debut:fin]}</doc>{doc[fin:]}"
        #print('yes')

    return str(phrase)
```

## Mise en forme des résultats sous forme de fichiers excels

```python
import re
import csv
```

```python
texte= lire_fichier(Path("annotation2fichiers/fichiers_annotees/2020-annote.txt"))

p= texte.split("\n")
```

```python

with open("mef2020_result-annotee.csv", "w", newline='') as csvfile:
    
    fieldnames = ['phrase_id',"balise_lieu_obj","balise_lieu_loc","balise_act","balise_doc","balise_perc","balise_tps",'balise_dyn', "phrase"]
    
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
      

    for i in range(len(p)):
        if "<dyn" in p[i]:
            
            #extraction des balises grâce aux regex
            extraction_dyn= re.findall("(?<=\<dyn>).*(?=<\/dyn>)", p[i])

            extraction_lieu_loc1= re.findall('(?<=\<L type="loc">).*?(?=<\/L>)', p[i])
            extraction_lieu_loc2= re.findall('(?<=\<L type=""loc"">).*?(?=<\/L>)', p[i])
            extraction_lieu_loc= [*extraction_lieu_loc2, *extraction_lieu_loc1]
            
            extraction_lieu_obj1= re.findall('(?<=\<L type="obj">).*?(?=<\/L>)', p[i])
            extraction_lieu_obj2= re.findall('(?<=\<L type=""obj"">).*?(?=<\/L>)', p[i])
            extraction_lieu_obj= [*extraction_lieu_obj1,*extraction_lieu_obj2]

            extraction_acteur= re.findall('(?<=\<A>).*?(?=<\/A>)', p[i])
            
            extraction_doc= re.findall("(?<=\<doc>).*(?=<\/doc>)", p[i])
            
            extraction_perc= re.findall("(?<=\<doc>).*(?=<\/doc>)", p[i])
            
            extraction_temps= re.findall('<TIMEX3.*?>.*?</TIMEX3', p[i])
            
            extraction_tps_net= re.findall('(?<=>).*?(?=<\/TIMEX3)', " ".join(extraction_temps))
     
            id_phrase= i
            
            phrase= re.sub("</?.*?>","",p[i])
            
            #nettoyage des balises 
            
            lieu_loc= re.sub("</?.*?>","","\n".join(extraction_lieu_loc))
            lieu_obj= re.sub("</?.*?>","","\n".join(extraction_lieu_obj))
            dyn= re.sub("</?.*?>","","\n".join(extraction_dyn))
            acteur= re.sub("</?.*?>","","\n".join(extraction_acteur))
            doc= re.sub("</?.*?>","","\n".join(extraction_doc))
            perc= re.sub("</?.*?>","","\n".join(extraction_perc))
            temps= re.sub("</?.*?>","","\n".join(extraction_tps_net))
        
            #écriture du document csv 
            writer.writerow({'phrase_id': id_phrase , 
                             'balise_dyn': dyn, 
                             'balise_lieu_obj': lieu_obj,
                             "balise_lieu_loc": lieu_loc,
                             "balise_act": acteur,
                             "balise_doc": doc,
                             "balise_perc": perc,
                             "balise_tps": temps,
                             "phrase": phrase
                            }
                           )
            
```
