```python
import re

def delete_html_tags(text):
    # Define regex pattern to match HTML tags
    pattern = r'<[^>]*>'
    # Use sub to replace all HTML tags with an empty string
    return re.sub(pattern, '', text)

# Example usage:
text_with_tags = "<p>This is <b>some</b> text with <a href='#'>HTML</a> tags.</p>"
clean_text = delete_html_tags(text_with_tags)
print(clean_text)  # Output: "This is some text with HTML tags."
```

```python
def compare_lists(nlieu, nlieugold):
    vrai_pos = 0
    faux_pos = 0
    faux_neg = 0

    for ele in nlieu:
        if any(ele in gold for gold in nlieugold):
            vrai_pos += 1
        else:
            faux_pos += 1

    for gold in nlieugold:
        if not any(gold in ele for ele in nlieu):
            faux_neg += 1

    return vrai_pos, faux_pos, faux_neg



vrai_pos, faux_pos, faux_neg = compare_lists(nlieu, nlieugold)
print("True positive:", vrai_pos)
print("False positive:", faux_pos)
print("False negative:", faux_neg)
```

```python
def calculate_metrics(vrai_pos, faux_pos, faux_neg):
    precision = vrai_pos / (vrai_pos + faux_pos)
    recall = vrai_pos / (vrai_pos + faux_neg)
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    return precision, recall, f1_score

# Example usage:
precision, recall, f1_score = calculate_metrics(vrai_pos, faux_pos, faux_neg)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1_score)
```

```python
from transformers import pipeline

def process_ner(line):
    ner = pipeline('token-classification', model='CATIE-AQ/NERmembert-base-3entities', tokenizer='CATIE-AQ/NERmembert-base-3entities', aggregation_strategy="simple")
    result = ner(line)  # Perform NER on the current line
    locs = [dico for dico in result if dico["entity_group"] == "LOC" ]  # Extract location entities
    if locs:
        for loc in locs:
            line = line.replace(loc["word"], f' <L>{loc["word"]}</L> ')  # Attach <L> symbol to each location word
    return line



```

```python
import re

def clean(file, outfile):
    with open(file, 'r', encoding="utf-8") as f:
        txt = f.read()
        txt = re.sub(r'(\w)([.,:?;!])', r'\1 \2', txt)
        txt = re.sub(r'([.,!?;:])(\w)', r'\1 \2', txt)
        #txt = re.sub("</L>’<L>", "’", txt)
        txt = re.sub("<L>'</L>","'",txt)
        txt = re.sub("</L>'<L>","'",txt)
        txt = re.sub(r'(>)([.,!?;:])', r'\1 \2', txt)
        txt = re.sub(r'<L>(.{1,2})</L>', lambda x: x.group(1), txt)
        txt = re.sub(r'\(<', r'( <', txt)
        txt = re.sub(r'>\)', r'> )', txt)

    # Write the modified text back to the same file
    with open(outfile, 'w', encoding="utf-8") as f:
        f.write(txt)

# Example usage:
clean("2003.txt","2003cl.txt")
```

```python
def previous2(line):
    line = re.sub("«",",",line)
    line = re.sub("»",",",line)
    line = re.sub(r"\bd'(\w+)", r"d' \1", line)
    line = re.sub(r"(<dyn>)(\w+)", r"\1 \2", line)
    line = re.sub(r"\bl'(\w+)", r"l' \1", line)
    line = re.sub(r"(\w+)(</dyn>)", r"\1 \2", line)
    line = re.sub(r"(>)([,.;?!])", r"\1 \2", line)

    line = re.sub(r"\((\w+)", r"( \1", line)
    line = re.sub(r"(\w+)\)", r"\1 )", line)


    words = line.split()
    pos_tags = [nlp(word)[0].pos_ for word in words]

    for i in range(len(words)-1):
        if "<L>" in words[i]:
            # Check if the word before <L> meets the specified conditions
            prev_word_index = i-1
            while prev_word_index >= 0 and (words[prev_word_index] not in ["du","avec","de"] and pos_tags[prev_word_index] not in ["CCONJ", "PUNCT","VERB","SCONJ","ADV","ADJ"]) and not re.match(r'^\d+$', words[prev_word_index]) and "</L" not in (words[prev_word_index]) :
                prev_word_index -= 1

            # Remove <L> in front of the last validated word
            if prev_word_index >= -1:
                words[i] = re.sub("<L>"," ",words[i])
                words[prev_word_index+1] = f"<L>{words[prev_word_index+1]} "
        if "</L>" in words[i]:
            # Check if the word after </L> is an adjective
            next_word_index = i + 1
            if next_word_index < len(words) and pos_tags[next_word_index] == "ADJ":
                # Remove </L> from the original word and add it at the end of the next word
                words[i] = re.sub("</L>", "", words[i])
                words[next_word_index] = f"{words[next_word_index]}</L> "

    return ' '.join(words)

# Example usage:
line = """
Je propose d'exploiter la petite marée du nord de l' <L>Adriatique</L>  pour <dyn>créer une succession de bassins</dyn> dont les niveaux fluctuent avec le sac et le ressac .
 """
modified_line = previous2(line)
print(modified_line)
```

```python
import spacy
import re

# Charger le modèle de langue française de spaCy
nlp = spacy.load("fr_core_news_sm")

# Définir une fonction pour vérifier si une partie contient une préposition
def contains_preposition(text):
    doc = nlp(text)
    for token in doc:
        if token.pos_ == "ADP" and token.lemma_ != "pour":  # ADP est l'étiquette POS pour les prépositions dans spaCy
            return True
    return False

# Fonction pour nettoyer le contenu des balises
def clean_text(text):
    # Supprimer toutes les balises HTML ou XML
    clean = re.sub(r"<.*?>", "", text)
    clean = re.sub(r"\bd'", "", clean)
    clean = re.sub(r"\bde\b", "", clean)
    return clean

# Lire le contenu du fichier
file_path = "2003fin2.txt"
with open(file_path, "r", encoding="utf-8") as file:
    content = file.read()

# Trouver et traiter toutes les balises <L>...</L>
def update_l_tags(content):
    l_tags = re.findall(r"<L>(.*?)</L>", content, re.DOTALL)
    for tag in l_tags:
        # Nettoyer le contenu des balises
        cleaned_tag = clean_text(tag)
        # Diviser la balise nettoyée par des espaces et vérifier chaque partie
        parts = cleaned_tag.split()
        if any(contains_preposition(part) for part in parts):
            updated_tag = f'<L type="loc">{tag}</L>'
        else:
            updated_tag = f'<L type="obj">{tag}</L>'
        content = content.replace(f"<L>{tag}</L>", updated_tag)
    return content

# Mettre à jour le contenu avec les balises traitées
updated_content = update_l_tags(content)

# Enregistrer le contenu mis à jour dans un nouveau fichier
output_file_path = "2003_updated.txt"
with open(output_file_path, "w", encoding="utf-8") as file:
    file.write(updated_content)

output_file_path
```

```python
from transformers import pipeline

def process_ner(line):
    ner = pipeline('token-classification', model='CATIE-AQ/NERmembert-base-3entities', tokenizer='CATIE-AQ/NERmembert-base-3entities', aggregation_strategy="simple")
    result = ner(line)  # Perform NER on the current line
    locs = [dico for dico in result if dico["entity_group"] == "PER" or dico["entity_group"] == "ORG" ]  # Extract location entities
    if locs:
        for loc in locs:
            line = line.replace(loc["word"], f' <A>{loc["word"]}</A> ')  # Attach <L> symbol to each location word
    return line



```

```python
def process_line(line):

    new_line = ""
    new_line+=process_ner(line)



    return new_line





# Usage
line = """
Ainsi pour <dyn>le</dyn> plan de rénovation du Chiado à  Lisbonne, j’ai décidé de définir <dyn>le</dyn>s lignes généra<dyn>le</dyn>s, <dyn>le</dyn>s principes et <dyn>le</dyn> règ<dyn>le</dyn>ment, en laissant aux propriétaires <dyn>le</dyn> choix des architectes.
"""
process_line(line)
```

```python
def process_file(input_file, output_file):
    with open(input_file, 'r') as file:
        lines = file.readlines()

    output_lines = []

    for line in lines:
        processed_line = process_line(line)
        output_lines.append(processed_line)

    with open(output_file, 'w') as file:
        for line in output_lines:
            file.write(line)
```

```python
import re
import pandas as pd

def extract_content_between_L_tags(input_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        text = file.read()

    # Regex pattern to find content between <L ...> and </L>
    pattern = r'<A[^>]*>(.*?)</A>'
    matches = re.findall(pattern, text, re.DOTALL)

    return matches

def save_to_excel(contents, output_file):
    df = pd.DataFrame(contents, columns=['Content'])
    df.to_excel(output_file, index=False)

if __name__ == "__main__":
    input_file = '2005.txt'  # Change this to your input file path
    output_file = 'output.xlsx'  # Change this to your desired output file path

    contents = extract_content_between_L_tags(input_file)
    save_to_excel(contents, output_file)

    print(f"Extracted content has been saved to {output_file}")
```
